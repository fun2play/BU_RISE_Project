```python
import numpy as np
import numpy as np
import mne
import matplotlib.pyplot as plt
import pandas as pd
import csv
from pathlib import Path
import os

def clean_data(subjn, ctal, eegtype, runn):
    subjectfile = f'dataset2/sub-HUP{subjn}/ses-presurgery/ieeg/sub-HUP{subjn}_ses-presurgery_task-{ctal}_acq-{eegtype}_run-0{runn}_'
    edf = mne.io.read_raw_edf(f'{subjectfile}ieeg.edf')
    header = ','.join(edf.ch_names)
    np.savetxt(f'{subjn}_{ctal}_{runn}_ieeg.csv', edf.get_data().T, delimiter=',', header=header)
    df = pd.read_csv(f'{subjn}_{ctal}_{runn}_ieeg.csv')
    channelfile = pd.read_csv(f'{subjectfile}channels.tsv', sep='\t')
    channelfile = channelfile[channelfile['status'] != 'bad']
    df = df[channelfile['name'].tolist()]
    # df.plot()
    #to_edf(df, subjn, ctal, eegtype, runn)
    
    data = df.values.T
    sfreq = float(channelfile['sampling_frequency'].iloc[0])
    ch_names = list(df.columns)
    ch_types = ['eeg'] * len(ch_names)
    info = mne.create_info(ch_names=ch_names, sfreq=sfreq, ch_types=ch_types)
    edf = mne.io.RawArray(data, info)
    mne.export.export_raw(f'cleanedeeg/{subjn}_{ctal}_{runn}_ieeg.edf', edf, fmt='edf', overwrite=True)
    #plt.show()

subjectlist = [ '060', '064', '065', '070', '074', '075', '080', '082', '086', '087', '088', '089', '094', '097',
               '105', '106', '107','111', '112', '114', '116', '117', '123','126', '130','133', '134', '135', 
               '138', '139', '140', '141', '142', '144', '146', '148', '150', '151', '157', '158', '160',
               '162', '163', '164', '165', '166', '171', '173', '173', '177', '179', '180', '181', '185', '187',
               '188', '190']

for subject in subjectlist:
    for run in [1, 2, 3, 4, 5]:
            if (os.path.exists(f'dataset2/sub-HUP{subject}/ses-presurgery/ieeg/sub-HUP{subject}_ses-presurgery_task-ictal_acq-seeg_run-0{run}_ieeg.edf')):
                #dataset2/sub-HUP144/ses-presurgery/ieeg/sub-HUP144_ses-presurgery_task-ictal_acq-seeg_run-01_
                try:
                    clean_data(subject, 'ictal', 'seeg', run)
                except:
                    continue
            if (os.path.exists(f'dataset2/sub-HUP{subject}/ses-presurgery/ieeg/sub-HUP{subject}_ses-presurgery_task-interictal_acq-seeg_run-0{run}_ieeg.edf')):
                try:
                    clean_data(subject, 'interictal', 'seeg', run)
                except:
                    continue
            if (os.path.exists(f'dataset2/sub-HUP{subject}/ses-presurgery/ieeg/sub-HUP{subject}_ses-presurgery_task-ictal_acq-ecog_run-0{run}_ieeg.edf')):
                try:
                    clean_data(subject, 'ictal', 'ecog', run)
                except:
                    continue
            if (os.path.exists(f'dataset2/sub-HUP{subject}/ses-presurgery/ieeg/sub-HUP{subject}_ses-presurgery_task-interictal_acq-ecog_run-0{run}_ieeg.edf')):
                try:
                   clean_data(subject, 'interictal', 'ecog', run)
                except:
                    continue
        

```
